```python
from sklearn import model_selection, preprocessing, naive_bayes
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import os
import re
import numpy as np
```

```python
datas = list()
labels = list()
for folder in os.listdir('data/'):
    label=folder
    folder = 'data/' + folder
    for each in os.listdir(folder):
        data_path = folder + "/" + each
        data = open(data_path,encoding='utf-8',errors='ignore').read()
        datas.append(data)
        labels.append(label)
        #  data = re.sub(',',' ',data)
        #with open('rawdata.csv','a',encoding='utf-8',errors='ignore') as f:
        #   f.write(data + ',' + folder + '\n')
print(labels)
```

```python
dictionaryLabels={'Bank': '0', 'Economy':'1', 'Education':'2', 'Entertainment': '3', 'National News':'4', 'Sports':'5', 'Technology':'6', 'Tourism':'7'}
    
```

```python
new_labels = list()
for eachLabel in labels:
    if eachLabel in dictionaryLabels:
         new_labels.append(dictionaryLabels[eachLabel])
#new_labels.append[dictionaryLables[eachlabel for eachlabel in labels if label in dictionaryLables]]
print(new_labels)
print('No of Datas',len(datas))
print('No of Labels',len(new_labels))
```

```python
#creating datafrmae in pandas
rawdata=pd.DataFrame()
rawdata['news']=datas
rawdata['labels']=new_labels
print(rawdata['news'])
```

```python
rawdata.tail()
```

```python
#shuffle dataset
#rawdata=rawdata.sample(frac=1)
```

```python
#splitting dataset into test and train data
train_x,valid_x,train_y,valid_y=model_selection.train_test_split(rawdata['news'],rawdata['labels'])
```

```python
#extracting feature using tfidf vector
tfidf_vect=TfidfVectorizer(analyzer='word',tokenizer=lambda x:x.split())
tfidf_vect.fit(train_x)
tfidf_vect.fit(valid_x)
xtrain=tfidf_vect.transform(train_x)
xvalid=tfidf_vect.transform(valid_x)
```

```python
classifier=naive_bayes.MultinomialNB(0.0001)
classifier.fit(xtrain,train_y)
```

```python
#prediction
prediction=classifier.predict(xvalid)
a=np.mean(prediction==valid_y)
```

```python
#removing punctuatuin
table=str.maketrans('','','!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~')
train_x=[i.translate(table) for i in train_x]
```

```python
def word_t(x):
    return x.split()
word_t("     ")
```

```python
#removing stopwords
import nltk
from nltk.tokenize import word_tokenize
def cleaner(x):
    x= [i for i in word_t(x) if i not in stop]
    x= ' '.join(x)
    return x
```

```python
print(train_x[0], "\n"*2)
train_x=[cleaner(i) for i in train_x]
valid_x=[i.translate(table) for i in valid_x]
valid_x=[cleaner(i) for i in valid_x]
print(train_x[0])
```

```python
import re

#pattern="$|$|$|$|$|$|$|$|$|$"
#url="           "
#urls = re.sub(pattern, '', url)
#print(urls, '\n')
def stemming(x):
    y=[]
    pattern="$|$|$|$|$|$|$|$|$|$|$|$|$|$|$|$|$|$|$|$"
   # url="           "
    for i in word_t(x):
        y.append(re.sub(pattern, '', i))
    result=" ".join(y)
    return result
        
   
    
stemming("           ")
    

```
